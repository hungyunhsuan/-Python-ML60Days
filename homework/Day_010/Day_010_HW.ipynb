題目:<br>
讀取STOCK_DAY_0050_202009.csv串聯STOCK_DAY_0050_202010.csv，並且找出open小於close的資料


```python
import pandas as pd

#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
file_202009 =  pd.read_csv('STOCK_DAY_0050_202009.csv')
file_202010 =  pd.read_csv('STOCK_DAY_0050_202010.csv')


#串聯
fileappend = file_202009.append(file_202010)


#找出open小於close的資料
open_min_close =  fileappend.loc[fileappend.open<fileappend.close]

open_min_close.set_index('date')
```
